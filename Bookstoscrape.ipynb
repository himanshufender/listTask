# Multiple pages 

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
book_details = []

allpages = ["http://books.toscrape.com/catalogue/page-{}.html".format(i)
            for i in range(1,51)]
column_names = ['Title', 'Link', 'Price', 'Rating', 'Stock']

book_Details = []

for url in allpages : 
    response = requests.get(url)
    soup = BeautifulSoup(response.text,'html.parser')
    
    books = soup.findAll('article', {'class':'product_pod'})
    
    for i in books: 
        title = i.h3.a['title']
        print(title)
        link = 'https://books.toscrape.com/catalogue/' + i.h3.a['href']
        print(link)

        priceText = i.find('p',{'class':'price_color'}).string
        pattern = r"([£$€])(\d+(?:\.\d+)?)"
        match = re.search(pattern, priceText) 
        price = float(match.group(2))
        print(price)

        rating = i.find('p', attrs={"class": "star-rating"})['class'][1]
        all_rating = {"One": 1, "Two": 2, "Three": 3, "Four": 4, "Five": 5}
        numRating = all_rating[rating]
        print(numRating)
        
        stock = i.find('p',{'class':'instock availability'}).text.strip()
        print(stock)
    
        book_details.append([title, link, price, numRating, stock])
        print(title, link, price,numRating, stock)
  
```

```python
allBooks = pd.DataFrame(book_details, columns = column_names)
```
